```python
import requests
from bs4 import BeautifulSoup
import os
from tqdm import tqdm

# 从Watkins数据集下载三种哺乳动物Bottlenose Dolphin、WhiteBeaked Dolphin和AtalnticSpotted Dolphin
# 可修改两个位置
# 1、url 设置要提取的Watkins的网页
# 2、os.makedirs和os.path.join 修改文件保存位置

# 目标网页 URL
# WhiltBeaked Dolphin
#url = "https://whoicf2.whoi.edu/science/B/whalesounds/fullCuts.cfm?SP=BD6B&YR=59"
#url = "https://whoicf2.whoi.edu/science/B/whalesounds/fullCuts.cfm?SP=BD6B&YR=62"
#url = "https://whoicf2.whoi.edu/science/B/whalesounds/fullCuts.cfm?SP=BD6B&YR=70"
#url = "https://whoicf2.whoi.edu/science/B/whalesounds/fullCuts.cfm?SP=BD6B&YR=72"
# Best
# url = "https://whoicf2.whoi.edu/science/B/whalesounds/bestOf.cfm?code=BD6B"

# AtlanticSpotted Dolphin
#url = "https://whoicf2.whoi.edu/science/B/whalesounds/fullCuts.cfm?SP=BD15F&YR=61"
#url = "https://whoicf2.whoi.edu/science/B/whalesounds/fullCuts.cfm?SP=BD15F&YR=72"
# Best
#url = "https://whoicf2.whoi.edu/science/B/whalesounds/bestOf.cfm?code=BD15F"

# Bottlenose Dolphin
#url = "https://whoicf2.whoi.edu/science/B/whalesounds/fullCuts.cfm?SP=BD19D&YR=94"
#url = "https://whoicf2.whoi.edu/science/B/whalesounds/fullCuts.cfm?SP=BD19D&YR=89"
#url = "https://whoicf2.whoi.edu/science/B/whalesounds/fullCuts.cfm?SP=BD19D&YR=84"
#url = "https://whoicf2.whoi.edu/science/B/whalesounds/fullCuts.cfm?SP=BD19D&YR=51"

# Best White-Sided Dolphin
url = "https://whoicf2.whoi.edu/science/B/whalesounds/bestOf.cfm?code=BD6A"

base_url = "https://whoicf2.whoi.edu"


# 发送请求获取网页内容
response = requests.get(url)
if response.status_code != 200:
    print("无法访问网页")
    exit()

# 解析 HTML
soup = BeautifulSoup(response.text, 'html.parser')

# 找到所有的音频链接
audio_links = []
for link in soup.find_all('a', href=True):
    href = link['href']
    if href.endswith(".wav"):
        audio_links.append(base_url + href)


#print(audio_links)  # 输出所有音频链接

# 创建保存音频的文件夹
os.makedirs(r"D:\Dataset\Data_Watkins\Best_WhiteSided_sounds", exist_ok=True)

# 下载所有音频文件
for audio_url in tqdm(audio_links, desc="Downloading", unit="file"):
    file_name = os.path.join(r"D:\Dataset\Data_Watkins\Best_WhiteSided_sounds", audio_url.split("/")[-1])
    audio_data = requests.get(audio_url, stream=True)
    
    with open(file_name, "wb") as file:
        for chunk in audio_data.iter_content(chunk_size=1024):
            if chunk:
                file.write(chunk)
    
    print(f"Saved: {file_name}")

print("所有音频文件已下载！")
```
